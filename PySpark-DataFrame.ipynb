# mean

```python
# add imputation cols to df
# imputer.fit(dfs).transform(dfs).show()
```

```python
# find the mean of 'salary' save to variable
mean_salary = dfs.agg({"Salary": "mean"}).collect()[0][0]
```

```python
# fill null with mean_salary value, in that specific column
dfs = dfs.fillna(mean_salary, subset=["Salary"])
```

```python
# MEAN FILLED FOR SALARY
dfs.show()
```

# Method 2

```python
from pyspark.ml.feature import Imputer

# Create an instance of the Imputer class with mean imputation
imputer_mean = Imputer(inputCols=["Age"], outputCols=["Age_imputed"], strategy="mean")

# Fit the imputer to the data and transform the data
imputed_df_mean = imputer_mean.fit(dfs).transform(dfs)

# Create an instance of the Imputer class with median imputation
imputer_median = Imputer(inputCols=["Age"], outputCols=["Age_imputed"], strategy="median")

# Fit the imputer to the data and transform the data
imputed_df_median = imputer_median.fit(dfs).transform(dfs)

# Create an instance of the Imputer class with mode imputation
imputer_mode = Imputer(inputCols=["Age"], outputCols=["Age_imputed"], strategy="mode")

# Fit the imputer to the data and transform the data
imputed_df_mode = imputer_mode.fit(dfs).transform(dfs)
```

```python
# new column
imputed_df_median.show(),imputed_df_mode.show()
```
